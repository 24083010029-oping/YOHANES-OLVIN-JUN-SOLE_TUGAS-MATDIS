```python
import matplotlib.pyplot as plt

# Fungsi deret harmonik
def harmonic_sum(n):
    return sum(1/i for i in range(1, n+1))

# Batas n yang mau dicek
N = 20

print("n | H_n | H_n < 2 ?")
print("="*35)
for n in range(1, N+1):
    Hn = harmonic_sum(n)
    print(f"n = {n:2d} | H_n = {Hn:.6f} | H_n < 2 ? {Hn < 2}")

# Cari n terakhir sebelum melewati 2
n = 1
while harmonic_sum(n) < 2:
    n += 1

n_max = n-1
print("\nRingkasan:")
print(f"Nilai n maksimal sebelum H_n melewati 2 adalah n = {n_max}")
print(f"Jumlah deret H_{n_max} = {harmonic_sum(n_max)}")

# Buat grafik
x = list(range(1, N+1))
y = [harmonic_sum(i) for i in x]

plt.figure(figsize=(9,5))
plt.plot(x, y, marker='o', label="Deret Harmonik H_n")
plt.axhline(2, color='r', linestyle='--', label="y = 2")
plt.axvline(n_max, color='g', linestyle=':', label=f"n={n_max}")
plt.title("Pertumbuhan Deret Harmonik")
plt.xlabel("n")
plt.ylabel("H_n")
plt.legend()
plt.grid(True)
plt.show()
```
